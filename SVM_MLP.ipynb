#### Regresion con redes neuronales para predecir el precio de casas de Boston

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from keras.datasets import boston_housing
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
```

```
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
```

```
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```
# escalar los datos para que tengan media = 0 y desviacion estandar = 1
scaler = StandardScaler().fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```
clf_1 = SVR()
```

```
clf_1.fit(x_train, y_train)
```

```
predicciones_1 = clf_1.predict(x_test)
```

```
predicciones_1
```

```
clf_2 = MLPRegressor()
```

```
clf_2.fit(x_train, y_train)
```

```
predicciones_2 = clf_2.predict(x_test)
```

```
predicciones_2
```

```
plt.plot(y_test)
plt.plot(predicciones_1)
plt.grid()
plt.xlabel('N_casa')
plt.ylabel('Precio')
plt.legend(['y_test', 'predicciones_SVM'])
plt.show()
```

```
plt.plot(y_test)
plt.plot(predicciones_2)
plt.grid()
plt.xlabel('N_casa')
plt.ylabel('Precio')
plt.legend(['y_test', 'predicciones_MLP'])
plt.show()
```
